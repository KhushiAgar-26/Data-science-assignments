```python
# Standardization of Numerical Variables
# ‚óè	Objective: To scale numerical variables for uniformity, improving the dataset‚Äôs suitability for analytical models.
# ‚óè	Steps:
# 	Explain the concept of standardization (z-score normalization).
# 	Standardize the numerical columns using the formula: z=x-mu/sigma
	
# 	Show before and after comparisons of the data distributions.
```

```python

# Select numerical variables for standardization
num_vars = ['Volume', 'Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']
# Before Standardization: Plot histograms or boxplots
plt.figure(figsize=(25,10))
for i, var in enumerate(num_vars, 1):
    plt.subplot(2, len(num_vars), i)
    data[var].hist()
    plt.title(f'Before Standardization\n{var}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')


```

```python
# Standardize numerical columns
for var in num_vars:
    mean = data[var].mean()
    std = data[var].std()
    data[var + '_standardized'] = (data[var] - mean) / std
```

```python
# After Standardization: Plot histograms or boxplots
plt.figure(figsize=(20,10))
for i, var in enumerate(num_vars, 1):
    plt.subplot(2, len(num_vars), len(num_vars) + i)
    data[var + '_standardized'].hist()
    plt.title(f'After Standardization\n{var}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
